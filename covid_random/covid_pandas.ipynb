```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

```python
#url to dataset in csv form
url_data = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
```

```python
#Lets grab the data
covid_df = pd.read_csv(url_data)
```

```python
#We are going to cut some data from the df as it isn't usefull here
covid_df.drop(['Province/State', 'Lat', 'Long'], axis=1, inplace=True)
```

```python
#There are multiple rows for the same country, lets sum these together
covid_df = covid_df.groupby('Country/Region').agg('sum')
```

```python
#Now let's plot for france

country = 'France'

ax = covid_df.loc[country].plot()
ax.set(ylabel="Number of confirmed cases")
ax = covid_df.loc[country].diff().plot(title='Analysis of' + ' ' + country,secondary_y=True, style='g')
ax.set(ylabel="d/dx")
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(covid_df, alpha=0.2)
```

```python
from pandas.plotting import lag_plot

plt.figure()

lag_plot(covid_df)
```

```python
from pandas.plotting import autocorrelation_plot

plt.figure()

autocorrelation_plot(covid_df)
```
